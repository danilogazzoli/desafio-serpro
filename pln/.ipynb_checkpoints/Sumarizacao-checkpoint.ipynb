```python
from urllib.request import Request, urlopen
import csv
from collections import defaultdict
from bs4 import BeautifulSoup
from lxml import etree
from pathlib import Path
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from string import punctuation
from nltk.probability import FreqDist
from heapq import nlargest
```

```python
columns = defaultdict(list) # each value in each column is appended to a list

with open(str(Path.cwd()) + '/desafio-pln-noticias.csv', newline='') as f:        
    reader = csv.DictReader(f, delimiter='|')
    for row in reader: 
        for (k,v) in row.items(): 
            columns[k].append(v) 

for row in columns['Link da notícia']:
    print(row)
```

```python
#request das páginas
paginas = []
for row in columns['Link da notícia']:
    link = Request(row, headers={'User-Agent':'Mozilla/5.0'})
    paginas.append(urlopen(link).read().decode('utf-8', 'ignore'))
```

```python
#raspagagem do texto nas páginas
noticias = []
for pagina in paginas:
    soup = BeautifulSoup(pagina, "html.parser")
    divs = soup.find_all("div", {"class": "bbc-19j92fr"})
    texto_ = soup.find(id="content").text + '\n'
    for div in divs:
        texto_ = texto_ + div.text
    noticias.append(texto_)    
```

```python
from nltk.corpus import stopwords
stopwords = set(stopwords.words('portuguese') + list(punctuation))

for texto in noticias:
    sentencas = sent_tokenize(texto)
    palavras = word_tokenize(texto_.lower())
    palavras_sem_sw = [palavra for palavra in palavras if palavra not in stopwords]
    frequencia = FreqDist(palavras_sem_sw)
    sentencas_importantes = defaultdict(int)
    for i, sentenca in enumerate(sentencas):
        for palavra in word_tokenize(sentenca.lower()):
            if palavra in frequencia:
                sentencas_importantes[i] += frequencia[palavra]
    
    for i in sorted(sentencas_importantes):
        print(sentencas[i])                
    print('*'*100)    
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
import heapq
import spacy.cli
from itertools import zip_longest

spacy.cli.download("pt_core_news_sm")
```

```python
from spacy import displacy
from IPython.display import SVG, display
```

```python
nlp = spacy.load('pt_core_news_sm')
```

```python
def summarize(text, per):
    doc= nlp(text)
    tokens=[token.text for token in doc] #tokens no texto
    word_frequencies={} #frequência de palavras
    for word in doc:
        if word.text.lower() not in list(STOP_WORDS):
            if word.text.lower() not in punctuation:
                if word.text not in word_frequencies.keys():
                    word_frequencies[word.text] = 1
                else:
                    word_frequencies[word.text] += 1
    max_frequency=max(word_frequencies.values())
    for word in word_frequencies.keys():
        word_frequencies[word]=word_frequencies[word]/max_frequency
    sentence_tokens= [sent for sent in doc.sents]
    sentence_scores = {}
    for sent in sentence_tokens:
        for word in sent:
            if word.text.lower() in word_frequencies.keys():
                if sent not in sentence_scores.keys():                            
                    sentence_scores[sent]=word_frequencies[word.text.lower()]
                else:
                    sentence_scores[sent]+=word_frequencies[word.text.lower()]
    select_length=int(len(sentence_tokens)*per)
    summary=nlargest(select_length, sentence_scores,key=sentence_scores.get)
    final_summary=[word.text for word in summary]
    summary=''.join(final_summary)
    return summary
```

```python
sumario = []
for noticia in noticias:
    sumario.append(summarize(noticia, 0.04))
```

```python
def showSVG(s):
    display(SVG(s))

for texto in sumario:
    print(texto)
    doc = nlp(texto)
    graph01 = displacy.render(doc)
    showSVG(graph01)
    print('*'*100)
```

```python
#cálculo de similaridade
similaridade = []
for count, value in enumerate(columns['Resumo referência']):
    resumo_oficial = nlp(value)
    resumo_calc = nlp(sumario[count])
    similar = resumo_calc.similarity(resumo_oficial) 
    print(f'A similaridade do resumo calculado com o oficial é: {similar}')
    similaridade.append(similar)

    
```

```python
d = [columns['Id da notícia'], sumario, similaridade]
export_data = zip_longest(*d, fillvalue = '')

with open('../resultado-pln-02639086901.csv', 'w', newline='') as csv_file:
    wr = csv.writer(csv_file, delimiter='|')
    wr.writerow(["ID da notícia","Sumário gerado", "Similaridade"])
    wr.writerows(export_data)    
                
csv_file.close()   
```
